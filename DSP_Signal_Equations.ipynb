<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Sine Wave:
</p>
    
$$
\Huge x(t) = A \sin(2\pi f t)
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Cosine Wave:
</p>
    
$$
\Huge x(t) = A \cos(2\pi f t)
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Ramp Wave:
</p>
    
$$
\Huge
x(t) = 
\begin{cases}
    t, & t \ge 0 \\
    0, & t < 0
\end{cases}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Exponential Pulse:
</p>

$$
\Huge
x(t) = R e^{at}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Impulse Pulse:
</p>

$$
\Huge
x(t) =
\begin{cases}
    A, & t = 0 \\
    0, & \text{otherwise}
\end{cases}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Rectangular Pulse:
</p>

$$
\Huge
x(t) =
\begin{cases}
    A, & |t| \le T \\
    0, & \text{otherwise}
\end{cases}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Bipolar Pulse:
</p>

$$
\Huge
x(t) =
\begin{cases}
 A, & 0 \le t < T \\
-A, & -T \le t < 0 \\
 0, & \text{otherwise}
\end{cases}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Triangular Pulse:
</p>

$$
\Huge
x(t) = A - 2 A \Bigg| \frac{t - T / 4}{T} - floor(\frac{t - T / 4}{T} + 0.5)\Bigg|
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Composite Wave:
</p>

$$
\Huge
x(t) = 10\sin(6280t) + 5\cos(3140t)
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Truncated Cosine Wave:
</p>

$$
\Huge
x(t) = 
\begin{cases}
    \cos(2\pi f t), & 0 \le t < 1 \\
    0, & \text{otherwise}
\end{cases}
$$

<hr>

<p style="font-family:'Montserrat', sans-serif; font-weight:600; text-transform: uppercase;">
Sinc Wave:
</p>

$$
\Huge
x(t) = \frac{\sin(2 \pi f t)}{t}  
$$
